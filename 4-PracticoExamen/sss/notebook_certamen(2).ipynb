```python
# Convierte nombres de columnas del DataFrame a Espanol
EXPECTED_SCHEMA_ES = {

    "age": "Edad",
    "gender": "Genero",
    "ethnicity": "Etnicidad",
    "education_level": "Nivel_educativo",
    "income_level": "Nivel_de_ingresos",
    "employment_status": "Estado_laboral",
    "smoking_status": "Estado_de_fumador",
    "alcohol_consumption_per_week": "Consumo_de_alcohol_por_semana",
    "physical_activity_minutes_per_week": "Minutos_de_actividad_fisica_por_semana",
    "diet_score": "Puntuacion_de_dieta",
    "sleep_hours_per_day": "Horas_de_suehnio_por_dia",
    "screen_time_hours_per_day": "Horas_de_tiempo_de_pantalla_por_dia",
    "family_history_diabetes": "Antecedentes_familiares_de_diabetes",
    "hypertension_history": "Historial_de_hipertension",
    "cardiovascular_history": "Historial_cardiovascular",
    "bmi": "Bmi",
    "waist_to_hip_ratio": "Relacion_cadera_cintura",
    "systolic_bp": "Presion_arterial_sistolica",
    "diastolic_bp": "Presion_arterial_diastolica",
    "heart_rate": "Frecuencia_cardiaca",
    "cholesterol_total": "Colesterol_total",
    "hdl_cholesterol": "Colesterol_hdl",
    "ldl_cholesterol": "Colesterol_ldl",
    "triglycerides": "Trigliceridos",
    "glucose_fasting": "Glucosa_en_ayuno",
    "glucose_postprandial": "Glucosa_postprandial",
    "insulin_level": "Nivel_de_insulina",
    "hba1c": "Hba1c",
    "diabetes_risk_score": "Puntuacion_de_riesgo_de_diabetes",
    "diabetes_stage": "Etapa_de_diabetes",
    "diagnosed_diabetes": "Diabete_diagnosticada",

}

# Establece la conversion de tipos de datos del DataFrame
EXPECTED_SCHEMA = {

    'Edad': 'Int64',
    'Genero': 'string',
    'Etnicidad': 'string',
    'Nivel_educativo': 'string',
    'Nivel_de_ingresos': 'string',
    'Estado_laboral': 'string',
    'Estado_de_fumador': 'string',
    'Consumo_de_alcohol_por_semana': 'Int64',
    'Minutos_de_actividad_fisica_por_semana': 'Int64',
    'Puntuacion_de_dieta': 'Int64',
    'Horas_de_suehnio_por_dia': 'Int64',              
    'Horas_de_tiempo_de_pantalla_por_dia': 'Int64',   
    'Antecedentes_familiares_de_diabetes': 'Int64',
    'Historial_de_hipertension': 'Int64',
    'Historial_cardiovascular': 'Int64',
    'Bmi': 'Float64',                                   
    'Relacion_cadera_cintura': 'Float64',                
    'Presion_arterial_sistolica': 'Int64',
    'Presion_arterial_diastolica': 'Int64',
    'Frecuencia_cardiaca': 'Int64',
    'Colesterol_total': 'Int64',
    'Colesterol_hdl': 'Int64',
    'Colesterol_ldl': 'Int64',
    'Trigliceridos': 'Int64',
    'Glucosa_en_ayuno': 'Int64',
    'Glucosa_postprandial': 'Int64',
    'Nivel_de_insulina': 'Float64',                      
    'Hba1c': 'Float64',                                  
    'Puntuacion_de_riesgo_de_diabetes': 'Int64',
    'Etapa_de_diabetes': 'string',
    'Diabete_diagnosticada': 'Int64'
}

# Se aseguran los tipos de datos segun el esquema hecho anteriormente
def aseguraTipoDatosDeseados(df: pd.DataFrame, schema: dict = EXPECTED_SCHEMA) -> pd.DataFrame:

    df = df.copy()

    # Quitar espacios y pasa las columnas a mayusculas
    df.columns = df.columns.str.strip().str.lower()

    # Renombra el DataFrame a espanol
    df = df.rename(columns=EXPECTED_SCHEMA_ES)


    for c, t in schema.items():

        if c in df.columns:
            try:
                if t == "Int64":
                    # Parseo de datos + redondeo 
                    df[c] = pd.to_numeric(df[c], errors="coerce").round().astype("Int64")

                elif t == "Float64":
                    # Mantiene los decimales para los tipos de datos Float
                    # Los decimales son importantes para estos datos en especifico
                    df[c] = pd.to_numeric(df[c], errors="coerce").astype("Float64")
                else:
                    df[c] = df[c].astype(t)
            except Exception:
                # fallback
                if t == "string":
                    df[c] = df[c].astype("string")
                else:
                    df[c] = pd.to_numeric(df[c], errors="coerce")

    return df
```

```python
import numpy as np
# Objetivo: Se aplico winsorizar pra acontrolar los outliers en las variables numericas del DataFrame.
# Esto se hizo para basado en el rango intercuartilico y se fijo un limite superior y un limite inferior.
# Cualquier valor por debajo del limite inferior, se reemplaza por el mismo limite inferior, lo mismo sucede con el limite superior
def winsorizarOutliers(df: pd.DataFrame, cols=None, k:float=3) -> pd.DataFrame:
    df_cast = df.copy()
    if cols is None:
        cols = df_cast.select_dtypes(include="number").columns.tolist()
    for col in cols:
        q1 = df_cast[col].quantile(0.25)
        q3 = df_cast[col].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - k*iqr
        upper = q3 + k*iqr
        df_cast[col] = np.where(df_cast[col] < lower, lower,
                          np.where(df_cast[col] > upper, upper, df_cast[col]))
    return df_cast
```

### **PCA:** Calculo de estandarizacion, tabla de valores y K

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Se preparan los datos numericos
df = df_scaled.copy()
target_col = "Diabete_diagnosticada"

# Se excluye la variable objetivo
X_num = df.drop(columns=[target_col], errors="ignore").select_dtypes(include=["number"]).copy()


X_num = X_num.apply(pd.to_numeric, errors="coerce")
X_num = X_num.fillna(X_num.median(numeric_only=True))

print(f"Variables numÃ©ricas usadas en PCA: {X_num.shape[1]}")

# Se realiza es escalado general
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num)

# Se aplica PCA
pca = PCA()  # calcula todas las componentes
X_pca_full = pca.fit_transform(X_scaled)

# Datos de varianza explicada
var_ratio = pca.explained_variance_ratio_         
var_pct = (var_ratio * 100).round(2)                
var_acum_pct = var_pct.cumsum().round(2)

# Crea la tabla con resultados
tabla_pca = pd.DataFrame({
    "PC_i": [f"PC{i+1}" for i in range(len(var_pct))],
    "varianza_i_%": var_pct,
    "acumulada_%": var_acum_pct
})


display(tabla_pca.head(15))

umbral = 0.93
K = int(np.argmax(pca.explained_variance_ratio_.cumsum() >= umbral) + 1)
var_acum_K = (pca.explained_variance_ratio_.cumsum()[K-1] * 100).round(2)

print(f"\nNÃºmero de componentes retenidos (K): {K}")
print(f"Varianza explicada acumulada con K={K}: {var_acum_K}%")

X_pca = X_pca_full[:, :K]
print(f"Matriz reducida: {X_pca.shape[0]} filas x {X_pca.shape[1]} componentes principales")
```

```python

import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neural_network import MLPClassifier

# --- Preparar X (solo numÃ©ricas) e y (0/1) ---
X_full = df_scaled.copy()

# Evitar fuga semÃ¡ntica si existe esta columna
X_full = X_full.drop(columns=[c for c in ["Etapa_de_diabetes"] if c in X_full.columns], errors="ignore")

y = pd.to_numeric(X_full["Diabete_diagnosticada"], errors="coerce")
X = X_full.drop(columns=["Diabete_diagnosticada"], errors="ignore").select_dtypes(include=["number"])

mask = y.isin([0, 1])
X = X.loc[mask].reset_index(drop=True)
y = y.loc[mask].astype(int).reset_index(drop=True)


X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)


mlp = MLPClassifier(
    hidden_layer_sizes=(32, 16),
    activation='relu',
    solver='adam',
    max_iter=500,
    random_state=42
)


param_grid = {'alpha': [0.0001, 0.001, 0.01, 0.1]}
grid = GridSearchCV(mlp, param_grid, cv=3, scoring='accuracy', n_jobs=-1)

grid.fit(X_train, y_train)  
best_mlp = grid.best_estimator_
```
