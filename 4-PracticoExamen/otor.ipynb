```python
import matplotlib.pyplot as plt
colsNum = df.select_dtypes(include="number").columns.tolist()
for col in colsNum:
    plt.boxplot(df[col], patch_artist=True)
    plt.title(f"Boxplot de {col}")
    plt.show()
```

```python
import numpy as np
def winsorizarOutliers(df: pd.DataFrame, cols=None, k:float=3) -> pd.DataFrame:
    df_clean = df.copy()
    if cols is None:
        cols = df_clean.select_dtypes(include="number").columns.tolist()
    for col in cols:
        q1 = df_clean[col].quantile(0.25)
        q3 = df_clean[col].quantile(0.75)
        iqr = q3 - q1
        lower = q1 - k * iqr
        upper = q3 + k * iqr
        df_clean[col] = np.where(df_clean[col] < lower, lower,
                          np.where(df_clean[col] > upper, upper, df_clean[col]))
    return df_clean
```

```python
df_clean =winsorizarOutliers(df)
```

```python
import matplotlib.pyplot as plt
colsNum = df_clean.select_dtypes(include="number").columns.tolist()
for col in colsNum:
    plt.boxplot(df_clean[col], patch_artist=True)
    plt.title(f"Boxplot de {col}")
    plt.show()
```

```python
#Aplicamos OneHotEncoder y OrdinalEncoder en las variables respectivas.
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
import pandas as pd 

colsOHE = ['Genero', 'Etnicidad', 'Estado_laboral', 'Estado_de_fumador']
colsORD = ['Nivel_educativo', 'Nivel_de_ingresos', 'Etapa_de_diabetes']

encoderOHE = OneHotEncoder(drop='first', sparse_output=False) #-> matriz densa para pandas
encoderORD = OrdinalEncoder()

preprocessor = ColumnTransformer(
    transformers=[
        ('onehot', encoderOHE, colsOHE), #Aplica OHE
        ('ordinal', encoderORD, colsORD) # Aplica OE
    ], 
    remainder='passthrough' #No toca las otras columnas
)

df_encoded = preprocessor.fit_transform(df_clean)

namesOHE = preprocessor.named_transformers_['onehot'].get_feature_names_out(colsOHE)

encodedCols = list(namesOHE) + colsORD + [
    col for col in df_clean.columns if col not
      in colsOHE + colsORD
    #Trae las que no se tocaron en el remainder
]

df_encoded = pd.DataFrame(df_encoded, columns=encodedCols)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

# ====== A) Seleccionar solo variables numéricas ======
# Excluir la variable objetivo si está
X_num = (df_encoded
         .drop(columns=['diagnosed_diabetes'], errors='ignore')
         .select_dtypes(include=['number'])
         .copy())

print(f"Variables numéricas seleccionadas: {X_num.shape[1]}")

# ====== B) Escalado ======
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num)

# ====== C) PCA ======
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# ====== D) Varianza explicada ======
varComp = pca.explained_variance_ratio_
varAcum = varComp.cumsum()

tablaPCA = pd.DataFrame({
    'PC_i': [f"PC{i+1}" for i in range(len(varComp))],
    'varianza_i': (varComp * 100).round(2),
    'acumulada': (varAcum * 100).round(2)
})

# ====== E) Elegir K ======
K = (varAcum >= 0.90).argmax() + 1
print(f"Número de componentes principales retenidas: K = {K}")
display(tablaPCA.head(15))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# ========= A) Preparar datos numéricos (sin la variable objetivo) =========
df = df_encoded.copy()
target_col = "Diabete_diagnosticada"

# Solo numéricas y excluimos la variable objetivo
X_num = (df.drop(columns=[target_col], errors="ignore")
           .select_dtypes(include=["number","Int64","Float64"])
           .copy())

# Seguridad: coerción numérica e imputación simple con mediana
X_num = X_num.apply(pd.to_numeric, errors="coerce")
X_num = X_num.fillna(X_num.median(numeric_only=True))

print(f"Variables numéricas usadas en PCA: {X_num.shape[1]}")

# ========= B) Estandarizar =========
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_num)

# ========= C) PCA (todas las componentes) =========
pca = PCA()                 # calcula todas las PCs disponibles
X_pca_full = pca.fit_transform(X_scaled)

# Varianzas explicadas (individual y acumulada)
var_ratio = pca.explained_variance_ratio_           # fracción (0..1)
var_pct = (var_ratio * 100).round(2)                # %
var_acum_pct = var_pct.cumsum().round(2)

# ========= D) Tabla solicitada: PC_i | varianza_i | acumulada =========
tabla_pca = pd.DataFrame({
    "PC_i": [f"PC{i+1}" for i in range(len(var_pct))],
    "varianza_i_%": var_pct,
    "acumulada_%": var_acum_pct
})

display(tabla_pca.head(15))   # muestra las primeras 15; puedes mostrar toda si quieres
# display(tabla_pca)          # <- descomenta para verla completa

# ========= E) Elegir K por umbral (ej. 95%) =========
umbral = 0.95  
K = int(np.argmax(pca.explained_variance_ratio_.cumsum() >= umbral) + 1)
var_acum_K = (pca.explained_variance_ratio_.cumsum()[K-1] * 100).round(2)

print(f"K retenidos para {int(umbral*100)}%: {K}")
print(f"Varianza explicada acumulada con K={K}: {var_acum_K}%")

# (Opcional) Dataset reducido para modelar
X_pca = X_pca_full[:, :K]
print("Matriz reducida (filas, K):", X_pca.shape)
```
