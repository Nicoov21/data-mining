```python
csvFile = "WorldCups.csv"
```

```python
import os
assert os.path.exists(csvFile), f"No existe el archivo"
```

```python
import pandas as pd
df = pd.read_csv(csvFile)
```

```python
import matplotlib.pyplot as plt

ye = df['Winner'].value_counts()

plt.figure(figsize=(12,6))
plt.bar(ye.index, ye.values)
plt.show()
```

```python
import matplotlib.pyplot as plt

year = df['Year']
goals = df['GoalsScored']

plt.figure(figsize=(12,6))
plt.plot(year, goals)
plt.title("Anotacin de goles al largo de aos")
plt.show()
```

```python
promGoles = df['GoalsScored'] / df['MatchesPlayed']
year = df['Year']

plt.figure(figsize=(12,6))
plt.plot(year, promGoles)
plt.title("Anotacin de goles al largo de aos")
plt.show()
```
