```python
import pandas as pd

df = pd.read_csv('Data.csv')
```

```python
df.describe(include='all') #Estadisticos descriptivos de todas las columnas
```

```python
df.isna().sum().sort_values(ascending=False) #Sumamos los nulos y los ordenamos de forma decreciente
```

```python
import missingno as msn
import matplotlib.pyplot as plt

msn.matrix(df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numCols = df.select_dtypes(include="number").columns.tolist()

for col in numCols:
    serie = pd.to_numeric(df[col], errors="coerce")
    sns.boxplot(serie)
    plt.title(f"Boxplot de {col}")
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numCols = df.select_dtypes(include="number").columns.tolist()

for col in numCols:
    serie = pd.to_numeric(df[col], errors="coerce")
    sns.histplot(serie, bins=5)
    plt.title(f"Boxplot de {col}")
    plt.show()
```
